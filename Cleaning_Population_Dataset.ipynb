```python
population_df['2011 Population'] = population_df['2011 Population'].astype('str')
population_df['2011 Population'] = population_df['2011 Population'].str.replace(',', '')
population_df['2011 Population'] = pd.to_numeric(population_df['2011 Population'], errors='coerce')

population_df['2012 Population'] = population_df['2012 Population'].astype('str')
population_df['2012 Population'] = population_df['2012 Population'].str.replace(',', '')
population_df['2012 Population'] = pd.to_numeric(population_df['2012 Population'], errors='coerce')

population_df['2013 Population'] = population_df['2013 Population'].astype('str')
population_df['2013 Population'] = population_df['2013 Population'].str.replace(',', '')
population_df['2013 Population'] = pd.to_numeric(population_df['2013 Population'], errors='coerce')

population_df['2014 Population'] = population_df['2014 Population'].astype('str')
population_df['2014 Population'] = population_df['2014 Population'].str.replace(',', '')
population_df['2014 Population'] = pd.to_numeric(population_df['2014 Population'], errors='coerce')

population_df['2015 Population'] = population_df['2015 Population'].astype('str')
population_df['2015 Population'] = population_df['2015 Population'].str.replace(',', '')
population_df['2015 Population'] = pd.to_numeric(population_df['2015 Population'], errors='coerce')

population_df['2016 Population'] = population_df['2016 Population'].astype('str')
population_df['2016 Population'] = population_df['2016 Population'].str.replace(',', '')
population_df['2016 Population'] = pd.to_numeric(population_df['2016 Population'], errors='coerce')

population_df['2017 Population'] = population_df['2017 Population'].astype('str')
population_df['2017 Population'] = population_df['2017 Population'].str.replace(',', '')
population_df['2017 Population'] = pd.to_numeric(population_df['2017 Population'], errors='coerce')

population_df['2018 Population'] = population_df['2018 Population'].astype('str')
population_df['2018 Population'] = population_df['2018 Population'].str.replace(',', '')
population_df['2018 Population'] = pd.to_numeric(population_df['2018 Population'], errors='coerce')

population_df['2019 Population'] = population_df['2019 Population'].astype('str')
population_df['2019 Population'] = population_df['2019 Population'].str.replace(',', '')
population_df['2019 Population'] = pd.to_numeric(population_df['2019 Population'], errors='coerce')

population_df['2020 Population'] = population_df['2020 Population'].astype('str')
population_df['2020 Population'] = population_df['2020 Population'].str.replace(',', '')
population_df['2020 Population'] = pd.to_numeric(population_df['2020 Population'], errors='coerce')

population_df['2021 Population'] = population_df['2021 Population'].astype('str')
population_df['2021 Population'] = population_df['2021 Population'].str.replace(',', '')
population_df['2021 Population'] = pd.to_numeric(population_df['2021 Population'], errors='coerce')
```

```python
def percentage_change(col1,col2):
    return ((col2 - col1) / col1) * 100
```

```python
population_df['2012 Population Percent Change'] = percentage_change(population_df['2011 Population'],population_df['2012 Population'])    
population_df['2013 Population Percent Change'] = percentage_change(population_df['2012 Population'],population_df['2013 Population'])
population_df['2014 Population Percent Change'] = percentage_change(population_df['2013 Population'],population_df['2014 Population'])
population_df['2015 Population Percent Change'] = percentage_change(population_df['2014 Population'],population_df['2015 Population'])
population_df['2016 Population Percent Change'] = percentage_change(population_df['2015 Population'],population_df['2016 Population'])
population_df['2017 Population Percent Change'] = percentage_change(population_df['2016 Population'],population_df['2017 Population'])
population_df['2018 Population Percent Change'] = percentage_change(population_df['2017 Population'],population_df['2018 Population'])
population_df['2019 Population Percent Change'] = percentage_change(population_df['2018 Population'],population_df['2019 Population'])
population_df['2020 Population Percent Change'] = percentage_change(population_df['2019 Population'],population_df['2020 Population'])
population_df['2021 Population Percent Change'] = percentage_change(population_df['2020 Population'],population_df['2021 Population'])
```

```python
def three_year_avg(col1,col2,col3):
    return ((col1 + col2 + col3) / 3)

def five_year_avg(col1,col2,col3,col4,col5):
    return ((col1 + col2 + col3 + col4 + col5) / 5)

def ten_year_avg(col1,col2,col3,col4,col5,col6,col7,col8,col9,col10):
    return ((col1 + col2 + col3 + col4 + col5 + col6 + col7 + col8 + col9 + col10) / 10)
```

```python
#add column for the last 3 year percent change average
population_df['3 Year Average Percent Change'] = three_year_avg(population_df['2019 Population Percent Change'],population_df['2020 Population Percent Change'],population_df['2021 Population Percent Change'])
#add column for the last 5 year percent change average
population_df['5 Year Average Percent Change'] = five_year_avg(population_df['2017 Population Percent Change'],population_df['2018 Population Percent Change'],population_df['2019 Population Percent Change'],population_df['2020 Population Percent Change'],population_df['2021 Population Percent Change'])
#add column for the last 10 year percent change average
population_df['10 Year Average Percent Change'] = ten_year_avg(population_df['2012 Population Percent Change'],population_df['2013 Population Percent Change'],population_df['2014 Population Percent Change'],population_df['2015 Population Percent Change'],population_df['2016 Population Percent Change'],population_df['2017 Population Percent Change'],population_df['2018 Population Percent Change'],population_df['2019 Population Percent Change'],population_df['2020 Population Percent Change'],population_df['2021 Population Percent Change'])
```
