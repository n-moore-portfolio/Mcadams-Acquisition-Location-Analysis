```python
#view data
firms_2017
```

```python
#filter out unecessary data from all dataframes
firms_2017_data = firms_2017[firms_2017["Meaning of Legal form of organization code"] == 'All establishments']
firms_2018_data = firms_2018[firms_2018["Meaning of Legal form of organization code"] == 'All establishments']
firms_2019_data = firms_2019[firms_2019["Meaning of Legal form of organization code"] == 'All establishments']
firms_2020_part_1_data = firms_2020_part_1[firms_2020_part_1["Meaning of Legal form of organization code"] == 'All establishments']
firms_2020_part_2_data = firms_2020_part_2[firms_2020_part_2["Meaning of Legal form of organization code"] == 'All establishments']
```

```python
#view data
firms_2020_part_1_data
```

```python
#concatenate the 2020 dataframes
firms_2020_data_df = pd.concat([firms_2020_part_1_data, firms_2020_part_2_data])
```

```python
#rename column header in the 2017 dataframe
firms_2017_df = firms_2017_data.rename({'Meaning of Employment size of establishments':'Meaning of Employment size of establishments code'}, axis=1)
```

```python
#concatenate all the yearly dataframes
firms_data = pd.concat([firms_2017_df, firms_2018_data, firms_2019_data, firms_2020_data_df])
firms_data
```

```python
#filter data
firms_data_filtered = firms_data[firms_data["Meaning of Employment size of establishments code"] == 'All establishments']
firms_data_filtered
```

```python
#filter data for big competitiors
establishment_list = ['Establishments with 100 to 249 employees','Establishments with 250 to 499 employees','Establishments with 500 to 999 employees','Establishments with 1,000 employees or more']
firms_data_a = firms_data[firms_data["Meaning of Legal form of organization code"] == 'All establishments']
firms_data_big_competitors = firms_data_a[firms_data_a["Meaning of Employment size of establishments code"].isin(establishment_list)]
firms_data_big_competitors.head(30)
```

```python
#Export clean data
firms_data.to_excel(r'C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\firms_data.xlsx')
firms_data_filtered.to_excel(r'C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\firms_data_filtered.xlsx')
firms_data_big_competitors.to_excel(r'C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\firms_data_big_competitors.xlsx')
```

```python
#filter specific code
firms_data_big_5413_competitors = firms_data_big_competitors[firms_data_big_competitors["2017 NAICS code"] == 5413]
firms_data_big_5413_competitors
```

```python
#create dataframe with big engineering competitors
firms_data_big_5413_competitors['Number of establishments'] = firms_data_big_5413_competitors['Number of establishments'].astype('str')
firms_data_big_5413_competitors['Number of establishments'] = firms_data_big_5413_competitors['Number of establishments'].str.replace(',', '')
firms_data_big_5413_competitors['Number of establishments'] = pd.to_numeric(firms_data_big_5413_competitors['Number of establishments'], errors='coerce')

industry = firms_data_big_5413_competitors.groupby(['Geographic Area Name','Year']).agg(sum_firms = ('Number of establishments','sum')).reset_index()
headers = ['Geographic Area Name','Year','sum_firms']
industry.columns = headers

industry_5413 = industry.pivot(index = ['Geographic Area Name'], columns = 'Year', values = ['sum_firms'])
industry_5413.columns = industry_5413.columns.droplevel()
modified_industry_5413=industry_5413.rename_axis(None,axis=1)
modified_industry_5413=modified_industry_5413.reset_index()

total_5413_firms = modified_industry_5413.rename({2017:'2017 Number of Firms',2018:'2018 Number of Firms',2019:'2019 Number of Firms',2020:'2020 Number of Firms'}, axis=1)
total_5413_firms
```
