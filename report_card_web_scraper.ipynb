```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
```

```python
#Create new web browser window
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
#Pull up the desired website to scrape
page_url = 'https://infrastructurereportcard.org/state-item/wisconsin/'
driver.get(page_url)
```

```python
#Create the soup
soup = bs(driver.page_source, "html.parser")
```

```python
#Create a list of column headers
table_headers = ['Infrastructure', 'Grade']
```

```python
#Scrape the content from the website
content = []
for x in soup.find_all('div', class_='e-flex'):
    div_tags = x.find_all('div', class_='e-grade')
    span_tags = x.find_all('span')
    span_val = [y.text for y in span_tags]
    content.append(span_val)
content
```

```python
#Create a dataframe of the content and column headers
infrastructure_df = pd.DataFrame(content, columns = table_headers)
infrastructure_df
```

```python
#Create a new column for the state names
infrastructure_df.insert(0, 'State', ['Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin','Wisconsin'])
infrastructure_df
```

```python
#Create a csv file for each state's information
infrastructure_df.to_csv(r"C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\wisconsin_infrastructure_data.csv", index=False)
```
