```python
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup as bs
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
page_url = 'https://www.usnews.com/news/best-states/rankings/natural-environment/pollution'
driver.get(page_url)
```

```python
time.sleep(5)

from selenium.webdriver.common.by import By

driver.find_element(By.XPATH, '//button[text()="Agree"]').click()
```

```python
time.sleep(5)

from selenium.webdriver.common.by import By

driver.find_element(By.XPATH, '//button/div[text()="Expand List"]').click()
```

```python
soup = bs(driver.page_source, "html.parser")
```

```python
table_head = soup.thead
```

```python
row_headers = []
for x in table_head.find_all('p'):
    row_headers.append(x.text)
row_headers
```

```python
table_body = soup.tbody
```

```python
rows = []
for x in table_body.find_all('tr')[1:]:
    td_tags = x.find_all('td')
    span_tags = x.find_all('span', class_="Span-sc-19wk4id-0 fgWqyH")
    span_val = [y.text for y in span_tags]
    rows.append(span_val)
rows
```

```python
states_df = pd.DataFrame(rows, columns = row_headers)
states_df['State'] = states_df['State'].replace(['WashingtonWashington', 'MinnesotaMinnesota','UtahUtah','New HampshireNew Hampshire','IdahoIdaho','NebraskaNebraska',
  'VirginiaVirginia','WisconsinWisconsin','MassachusettsMassachusetts','FloridaFlorida','VermontVermont','IowaIowa',
  'North CarolinaNorth Carolina','North DakotaNorth Dakota','South DakotaSouth Dakota','ColoradoColorado','MarylandMaryland',
  'GeorgiaGeorgia','New JerseyNew Jersey','ConnecticutConnecticut','New YorkNew York','OregonOregon','DelawareDelaware','CaliforniaCalifornia',
  'HawaiiHawaii','KansasKansas','MaineMaine','MissouriMissouri','TennesseeTennessee','IllinoisIllinois','TexasTexas',
  'IndianaIndiana','MontanaMontana','Rhode IslandRhode Island','WyomingWyoming','OhioOhio','NevadaNevada','MichiganMichigan',
  'ArizonaArizona','PennsylvaniaPennsylvania','KentuckyKentucky','South CarolinaSouth Carolina','OklahomaOklahoma','ArkansasArkansas',
  'AlaskaAlaska','AlabamaAlabama','West VirginiaWest Virginia','New MexicoNew Mexico','MississippiMississippi','LouisianaLouisiana'], ['Washington', 'Minnesota','Utah','New Hampshire','Idaho','Nebraska',
  'Virginia','Wisconsin','Massachusetts','Florida','Vermont','Iowa',
  'North Carolina','North Dakota','South Dakota','Colorado','Maryland',
  'Georgia','New Jersey','Connecticut','New York','Oregon','Delaware','California',
  'Hawaii','Kansas','Maine','Missouri','Tennessee','Illinois','Texas',
  'Indiana','Montana','Rhode Island','Wyoming','Ohio','Nevada','Michigan',
  'Arizona','Pennsylvania','Kentucky','South Carolina','Oklahoma','Arkansas',
  'Alaska','Alabama','West Virginia','New Mexico','Mississippi','Louisiana'])
states_df = states_df.mask(states_df.eq('None')).dropna()
```

```python
states_data_cleaned = states_df.reset_index(drop=True)
states_data_cleaned
```
