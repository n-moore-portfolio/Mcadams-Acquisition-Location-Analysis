```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
gdp = pd.read_excel(r"C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\population data\gdp.xlsx")
percent_change_gdp = pd.read_excel(r"C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\population data\gdp percents.xlsx")
industry_gdp = pd.read_excel(r"C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\population data\Table (1).xlsx")
industry_percent_change_gdp = pd.read_excel(r"C:\Users\Nathan\Documents\Python Projects\Mcadams Projects\Best Locations\data\population data\Table (2).xlsx")
```

```python
#view data
industry_gdp.head()
```

```python
#merge gdp dataframes
market_data = pd.merge(gdp,percent_change_gdp)
market_data.head()
```

```python
#merge industry dataframes
industry_market_data = pd.merge(industry_gdp, industry_percent_change_gdp)
industry_market_data.head()
```

```python
#merge dataframes
market_data_df = market_data.merge(industry_market_data,on='GeoFips')
market_data_df.head()
```

```python
#create new dataframe, rename columns, drop a column, view the data
clean_market_data = market_data_df.rename({'GeoName_x':'Region','2017_x':'2017 GDP','2018_x':'2018 GDP','2019_x':'2019 GDP','2020_x':'2020 GDP','2021_x':'2021 GDP','2016-2017_x':'2016-2017 GDP Percent Change','2017-2018_x':'2017-2018 GDP Percent Change','2018-2019_x':'2018-2019 GDP Percent Change','2019-2020_x':'2019-2020 GDP Percent Change','2020-2021_x':'2020-2021 GDP Percent Change','LineCode':'Line Code','Description':'Industry Description','2017_y':'2017 Industry GDP','2018_y':'2018 Industry GDP','2019_y':'2019 Industry GDP','2020_y':'2020 Industry GDP','2021_y':'2021 Industry GDP','2016-2017_y':'2016-2017 Industry GDP Percent Change','2017-2018_y':'2017-2018 Industry GDP Percent Change','2018-2019_y':'2018-2019 Industry GDP Percent Change','2019-2020_y':'2019-2020 Industry GDP Percent Change','2020-2021_y':'2020-2021 Industry GDP Percent Change'}, axis=1)
clean_market_data.drop('GeoName_y', axis=1, inplace=True)
clean_market_data.head()
```

```python
def three_year(col1,col2,col3):
    return ((col1 + col2 + col3) / 3)

def five_year(col1,col2,col3,col4,col5):
    return ((col1 + col2 + col3 + col4 + col5) / 5)
```

```python
#add column for the last 3 year gdp percent change average
clean_market_data['3 Year Average GDP Percent Change'] = three_year(clean_market_data['2018-2019 GDP Percent Change'],clean_market_data['2019-2020 GDP Percent Change'],clean_market_data['2020-2021 GDP Percent Change'])
#add column for the last 5 year gdp percent change average
clean_market_data['5 Year Average GDP Percent Change'] = five_year(clean_market_data['2016-2017 GDP Percent Change'],clean_market_data['2017-2018 GDP Percent Change'],clean_market_data['2018-2019 GDP Percent Change'],clean_market_data['2019-2020 GDP Percent Change'],clean_market_data['2020-2021 GDP Percent Change'])

#add column for the last 3 year industry gdp percent change average
clean_market_data['3 Year Average Industry GDP Percent Change'] = three_year(clean_market_data['2018-2019 Industry GDP Percent Change'],clean_market_data['2019-2020 Industry GDP Percent Change'],clean_market_data['2020-2021 Industry GDP Percent Change'])
#add column for the last 5 year industry gdp percent change average
clean_market_data['5 Year Average Industry GDP Percent Change'] = five_year(clean_market_data['2016-2017 Industry GDP Percent Change'],clean_market_data['2017-2018 Industry GDP Percent Change'],clean_market_data['2018-2019 Industry GDP Percent Change'],clean_market_data['2019-2020 Industry GDP Percent Change'],clean_market_data['2020-2021 Industry GDP Percent Change'])
```

```python
clean_market_data.head()
```

```python
#sort values
clean_market_data.sort_values(['Industry Description'], ascending=[True])
```

```python
#create a new dataframe for construction industry
construction_market_data = clean_market_data[clean_market_data["Line Code"] == 11]
construction_market_data.tail()
```

```python
#create a new dataframe for engineering industry
scientific_market_data = clean_market_data[clean_market_data["Line Code"] == 60]
scientific_market_data.tail()
```

```python
#create new dataframes for each specific industry's proportional gdp
scientific_proportions_data_df = scientific_market_data.filter(items=['Region','Line Code','Industry Description','2017 GDP','2018 GDP','2019 GDP','2020 GDP','2021 GDP','2017 Industry GDP','2018 Industry GDP','2019 Industry GDP','2020 Industry GDP','2021 Industry GDP','3 Year Average GDP Percent Change','5 Year Average GDP Percent Change','3 Year Average Industry GDP Percent Change','5 Year Average Industry GDP Percent Change'])

construction_proportions_data_df = construction_market_data.filter(items=['Region','Line Code','Industry Description','2017 GDP','2018 GDP','2019 GDP','2020 GDP','2021 GDP','2017 Industry GDP','2018 Industry GDP','2019 Industry GDP','2020 Industry GDP','2021 Industry GDP','3 Year Average GDP Percent Change','5 Year Average GDP Percent Change','3 Year Average Industry GDP Percent Change','5 Year Average Industry GDP Percent Change'])
```

```python
#function for new column
def gdp_proportion(col1,col2):
    return (col2/col1)
```

```python
#run data through function
scientific_proportions_data_df['2017 Industry Proportion of Whole GDP'] = gdp_proportion(scientific_market_data['2017 GDP'],scientific_market_data['2017 Industry GDP'])    
scientific_proportions_data_df['2018 Industry Proportion of Whole GDP'] = gdp_proportion(scientific_market_data['2018 GDP'],scientific_market_data['2018 Industry GDP'])
scientific_proportions_data_df['2019 Industry Proportion of Whole GDP'] = gdp_proportion(scientific_market_data['2019 GDP'],scientific_market_data['2019 Industry GDP'])
scientific_proportions_data_df['2020 Industry Proportion of Whole GDP'] = gdp_proportion(scientific_market_data['2020 GDP'],scientific_market_data['2020 Industry GDP'])
scientific_proportions_data_df['2021 Industry Proportion of Whole GDP'] = gdp_proportion(scientific_market_data['2021 GDP'],scientific_market_data['2021 Industry GDP'])

construction_proportions_data_df['2017 Industry Proportion of Whole GDP'] = gdp_proportion(construction_market_data['2017 GDP'],construction_market_data['2017 Industry GDP'])    
construction_proportions_data_df['2018 Industry Proportion of Whole GDP'] = gdp_proportion(construction_market_data['2018 GDP'],construction_market_data['2018 Industry GDP'])
construction_proportions_data_df['2019 Industry Proportion of Whole GDP'] = gdp_proportion(construction_market_data['2019 GDP'],construction_market_data['2019 Industry GDP'])
construction_proportions_data_df['2020 Industry Proportion of Whole GDP'] = gdp_proportion(construction_market_data['2020 GDP'],construction_market_data['2020 Industry GDP'])
construction_proportions_data_df['2021 Industry Proportion of Whole GDP'] = gdp_proportion(construction_market_data['2021 GDP'],construction_market_data['2021 Industry GDP'])
```

```python
scientific_proportions_data_df.head()
```

```python
#functions for new columns
def three_year_avg(col1,col2,col3):
    return ((col1 + col2 + col3) / 3)

def five_year_avg(col1,col2,col3,col4,col5):
    return ((col1 + col2 + col3 + col4 + col5) / 5)
```

```python
#run data through functions
scientific_proportions_data_df['Last 3 Year Avg Proportion'] = three_year_avg(scientific_proportions_data_df['2019 Industry Proportion of Whole GDP'],scientific_proportions_data_df['2020 Industry Proportion of Whole GDP'],scientific_proportions_data_df['2021 Industry Proportion of Whole GDP'])    
scientific_proportions_data_df['Last 5 Year Avg Proportion'] = five_year_avg(scientific_proportions_data_df['2017 Industry Proportion of Whole GDP'],scientific_proportions_data_df['2018 Industry Proportion of Whole GDP'],scientific_proportions_data_df['2019 Industry Proportion of Whole GDP'],scientific_proportions_data_df['2020 Industry Proportion of Whole GDP'],scientific_proportions_data_df['2021 Industry Proportion of Whole GDP'])

construction_proportions_data_df['Last 3 Year Avg Proportion'] = three_year_avg(construction_proportions_data_df['2019 Industry Proportion of Whole GDP'],construction_proportions_data_df['2020 Industry Proportion of Whole GDP'],construction_proportions_data_df['2021 Industry Proportion of Whole GDP'])    
construction_proportions_data_df['Last 5 Year Avg Proportion'] = five_year_avg(construction_proportions_data_df['2017 Industry Proportion of Whole GDP'],construction_proportions_data_df['2018 Industry Proportion of Whole GDP'],construction_proportions_data_df['2019 Industry Proportion of Whole GDP'],construction_proportions_data_df['2020 Industry Proportion of Whole GDP'],construction_proportions_data_df['2021 Industry Proportion of Whole GDP'])
```
